**Load data**

```python
df = sqlContext.read.format('csv')\
                    .option("header", "true")\
                    .option("inferschema", "true")\
                    .load('data/Crimes_-_2001_to_present.csv')
```

```python
print(df.count(), 'rows')
df.printSchema()
```

**(1) For the attributes  ‘Primary type’ and ‘Location description’, output the list of each value and the corresponding frequency count, sorted in descending order of the count, respectively.**

```python
df_PrimaryType = df.groupBy('Primary Type').count().orderBy('count', ascending=False).cache()
df_PrimaryType.show(df_PrimaryType.count())
```

```python
df_LocationDescription = df.groupBy('Location Description').count().orderBy('count', ascending=False).cache()
df_LocationDescription.show(df_LocationDescription.count())
```

**(2) Output the most frequently occurred ‘Primary type‘ for each possible value of ‘Location description’, sorted in descending order of the frequency count.**

```python
df_PrimaryType_of_LocationDescription = df.cube('Primary Type', 'Location Description').count()\
                                          .orderBy('Primary Type', 'count', 'Location Description', ascending=False).cache()
df_PrimaryType_of_LocationDescription.show(df_PrimaryType_of_LocationDescription.count())
```
